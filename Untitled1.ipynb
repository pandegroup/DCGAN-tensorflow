```python
import numpy as np
import tensorflow as tf
x = np.ones((10, 10,10)).astype(np.float32)
y = np.ones((10, 10)).astype(np.float32)
```

```python
tf.InteractiveSession()
```

```python
xt = tf.placeholder(tf.float32, shape=(10,10,10))
yt = tf.placeholder(tf.float32, shape=(10,10))

def mult(idx):
    #return(idx*2)
    return(tf.matmul(xt[tf.cast(idx, tf.int32)], yt))

elems = np.array([1., 2., 3., 4., 5., 6.]).astype(np.float32)
#tf.matmul(xt, yt).eval(feed_dict={xt: x, yt:y})
tf.map_fn(mult, elems).eval(feed_dict={xt: x, yt: y})
```
